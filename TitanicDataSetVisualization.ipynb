```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
survival_counts = df['Survived'].value_counts()

# Plot
plt.figure(figsize=(6,4))
survival_counts.plot(kind='bar', color=['red', 'green'])
plt.title('Survival Counts on the Titanic')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.xticks(ticks=[0,1], labels=['No', 'Yes'], rotation=0)
plt.show()
```

```python
survival_counts = df['Survived'].value_counts()

# Labels for the pie chart
labels = ['Did Not Survive', 'Survived']

# Colors for the slices
colors = ['red', 'green']

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(survival_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=[0.05, 0.05])
plt.title('Titanic Survival Distribution')
plt.show()
```

```python
survival_counts = df['Sex'].value_counts()

# Labels for the pie chart
labels = ['Male', 'Female']

# Colors for the slices
colors = ['red', 'green']

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(survival_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=[0.05, 0.05])
plt.title('Titanic Male-Female Distribution')
plt.show()
```

```python
survival_by_sex = df.groupby('Sex')['Survived'].mean()

plt.bar(survival_by_sex.index, survival_by_sex, color=['blue', 'pink'])
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.ylim(0,1)
plt.show()
```

```python
plt.hist([df[df['Pclass'] == 1]['Age'].dropna(),
          df[df['Pclass'] == 2]['Age'].dropna(),
          df[df['Pclass'] == 3]['Age'].dropna()],
         bins=30, stacked=True,
         label=['1st Class', '2nd Class', '3rd Class'])
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()

```

```python
survival_by_embark = df.groupby('Embarked')['Survived'].mean().dropna()

plt.bar(survival_by_embark.index, survival_by_embark, color='purple')
plt.title('Survival Rate by Embarkation Point')
plt.ylabel('Survival Rate')
plt.show()
```

```python
colors = df['Survived'].map({0: 'red', 1: 'green'})

plt.scatter(df['Age'], df['Fare'], c=colors, alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs Age Colored by Survival')
plt.show()
```
